````python
# åŸºæœ¬è–èª•æ¨¹å‡½æ•¸
def draw_christmas_tree(height):
    """
    ç¹ªè£½ä¸€å€‹ç°¡å–®çš„è–èª•æ¨¹
    
    åƒæ•¸:
    height (int): è–èª•æ¨¹çš„é«˜åº¦ï¼ˆå±¤æ•¸ï¼‰
    """
    print("ğŸ„ è–èª•æ¨¹ ğŸ„")
    print("=" * 20)
    
    # ç¹ªè£½æ¨¹å† 
    for i in range(height):
        # è¨ˆç®—ç©ºæ ¼æ•¸ï¼ˆè®“æ¨¹å±…ä¸­ï¼‰
        spaces = " " * (height - i - 1)
        # è¨ˆç®—æ˜Ÿè™Ÿæ•¸
        stars = "*" * (2 * i + 1)
        print(f"{spaces}{stars}")
    
    # ç¹ªè£½æ¨¹å¹¹
    trunk_spaces = " " * (height - 1)
    print(f"{trunk_spaces}|||")
    print(f"{trunk_spaces}|||")
    
    print("=" * 20)
    print("ğŸ… è–èª•å¿«æ¨‚ï¼ğŸ…")
````

````python
# æ¸¬è©¦åŸºæœ¬è–èª•æ¨¹
draw_christmas_tree(5)
````

````python
# é€²éšè–èª•æ¨¹å‡½æ•¸ - å¸¶è£é£¾
def draw_fancy_christmas_tree(height):
    """
    ç¹ªè£½ä¸€å€‹å¸¶è£é£¾çš„é€²éšè–èª•æ¨¹
    
    åƒæ•¸:
    height (int): è–èª•æ¨¹çš„é«˜åº¦ï¼ˆå±¤æ•¸ï¼‰
    """
    import random
    
    # è£é£¾ç¬¦è™Ÿ
    decorations = ["ğŸ„", "ğŸŒŸ", "ğŸ", "â„ï¸", "ğŸ””", "â­"]
    
    print("ğŸ„ è±ªè¯è–èª•æ¨¹ ğŸ„")
    print("=" * 25)
    
    # ç¹ªè£½æ¨¹å† 
    for i in range(height):
        spaces = " " * (height - i - 1)
        
        # å»ºç«‹æ¯ä¸€è¡Œçš„å…§å®¹
        line = ""
        for j in range(2 * i + 1):
            if j == 0 or j == 2 * i:  # é‚Šç·£ç”¨æ˜Ÿè™Ÿ
                line += "*"
            elif random.random() < 0.3:  # 30% æ©Ÿç‡æ”¾è£é£¾
                line += random.choice(decorations)
            else:
                line += "*"
        
        print(f"{spaces}{line}")
    
    # ç¹ªè£½æ¨¹å¹¹
    trunk_spaces = " " * (height - 2)
    print(f"{trunk_spaces}ğŸğŸğŸ")
    print(f"{trunk_spaces}ğŸğŸğŸ")
    
    print("=" * 25)
    print("ğŸ… è–èª•å¿«æ¨‚ï¼æ–°å¹´å¿«æ¨‚ï¼ğŸ…")
````

````python
# æ¸¬è©¦é€²éšè–èª•æ¨¹
draw_fancy_christmas_tree(6)
````

````python
# å½©è‰²è–èª•æ¨¹å‡½æ•¸ï¼ˆä½¿ç”¨ ANSI é¡è‰²ä»£ç¢¼ï¼‰
def draw_colored_christmas_tree(height):
    """
    ç¹ªè£½ä¸€å€‹å½©è‰²çš„è–èª•æ¨¹
    
    åƒæ•¸:
    height (int): è–èª•æ¨¹çš„é«˜åº¦ï¼ˆå±¤æ•¸ï¼‰
    """
    # ANSI é¡è‰²ä»£ç¢¼
    GREEN = "\033[92m"  # ç¶ è‰²
    RED = "\033[91m"    # ç´…è‰²
    YELLOW = "\033[93m" # é»ƒè‰²
    BLUE = "\033[94m"   # è—è‰²
    BROWN = "\033[33m"  # æ£•è‰²
    RESET = "\033[0m"   # é‡ç½®é¡è‰²
    
    print(f"{GREEN}ğŸ„ å½©è‰²è–èª•æ¨¹ ğŸ„{RESET}")
    print("=" * 25)
    
    # ç¹ªè£½æ¨¹å† 
    for i in range(height):
        spaces = " " * (height - i - 1)
        
        # å»ºç«‹å½©è‰²è¡Œ
        line = ""
        for j in range(2 * i + 1):
            if j == 0 or j == 2 * i:  # é‚Šç·£
                line += f"{GREEN}*{RESET}"
            elif j % 3 == 0:  # æ¯ä¸‰å€‹ä½ç½®æ”¾ç´…è‰²è£é£¾
                line += f"{RED}@{RESET}"
            elif j % 5 == 0:  # æ¯äº”å€‹ä½ç½®æ”¾é»ƒè‰²è£é£¾
                line += f"{YELLOW}*{RESET}"
            else:
                line += f"{GREEN}*{RESET}"
        
        print(f"{spaces}{line}")
    
    # ç¹ªè£½æ¨¹å¹¹
    trunk_spaces = " " * (height - 2)
    print(f"{trunk_spaces}{BROWN}|||{RESET}")
    print(f"{trunk_spaces}{BROWN}|||{RESET}")
    
    print("=" * 25)
    print(f"{BLUE}ğŸ… è–èª•å¿«æ¨‚ï¼ğŸ…{RESET}")
````

````python
# æ¸¬è©¦å½©è‰²è–èª•æ¨¹
draw_colored_christmas_tree(7)
````

````python
# äº’å‹•å¼è–èª•æ¨¹ç”Ÿæˆå™¨
def interactive_christmas_tree():
    """
    äº’å‹•å¼è–èª•æ¨¹ç”Ÿæˆå™¨
    """
    print("ğŸ„ æ­¡è¿ä½¿ç”¨è–èª•æ¨¹ç”Ÿæˆå™¨ï¼ğŸ„")
    print("=" * 35)
    
    while True:
        try:
            print("\nè«‹é¸æ“‡è–èª•æ¨¹é¡å‹ï¼š")
            print("1. åŸºæœ¬è–èª•æ¨¹")
            print("2. é€²éšè–èª•æ¨¹ï¼ˆå¸¶è£é£¾ï¼‰")
            print("3. å½©è‰²è–èª•æ¨¹")
            print("4. é€€å‡º")
            
            choice = input("\nè«‹è¼¸å…¥é¸é … (1-4): ").strip()
            
            if choice == "4":
                print("ğŸ… æ„Ÿè¬ä½¿ç”¨ï¼è–èª•å¿«æ¨‚ï¼ğŸ…")
                break
            
            if choice in ["1", "2", "3"]:
                height = int(input("è«‹è¼¸å…¥è–èª•æ¨¹é«˜åº¦ (3-10): "))
                
                if height < 3 or height > 10:
                    print("é«˜åº¦å¿…é ˆåœ¨ 3-10 ä¹‹é–“ï¼")
                    continue
                
                print("\n" + "=" * 40)
                
                if choice == "1":
                    draw_christmas_tree(height)
                elif choice == "2":
                    draw_fancy_christmas_tree(height)
                elif choice == "3":
                    draw_colored_christmas_tree(height)
                
                print("\n" + "=" * 40)
            else:
                print("ç„¡æ•ˆé¸é …ï¼Œè«‹é‡æ–°é¸æ“‡ï¼")
                
        except ValueError:
            print("è«‹è¼¸å…¥æœ‰æ•ˆçš„æ•¸å­—ï¼")
        except KeyboardInterrupt:
            print("\n\nğŸ… æ„Ÿè¬ä½¿ç”¨ï¼è–èª•å¿«æ¨‚ï¼ğŸ…")
            break
        except Exception as e:
            print(f"ç™¼ç”ŸéŒ¯èª¤ï¼š{e}")
````
